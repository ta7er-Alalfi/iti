```python
# Create a random NumPy array of 1000 normally distributed numbers using NumPy and save it to X


# Create a histogram plot of X
X = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(X, bins=30, edgecolor='black')

ax.set_title("Histogram of Normally Distributed Data")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.show()
```

Now let's try make some subplots. A subplot is another name for a figure with multiple plots on it.

```python
# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
fig, ax = plt.subplots(2, 2)

plt.show()
```

Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 5))

# Plot X versus X/2 on the top left axes
X = np.linspace(0, 10, 100)
ax[0, 0].plot(X, X/2)
ax[0, 0].set_title("X vs X/2")

# Scatter plot of 10 random numbers on each axis on the top right subplot
ax[0, 1].scatter(np.random.rand(10), np.random.rand(10))
ax[0, 1].set_title("Random Scatter Plot")

# Bar graph of the favourite food keys and values on the bottom left subplot
favorite_foods = {"Pizza": 120, "Sushi": 200, "Burger": 80}
ax[1, 0].bar(favorite_foods.keys(), favorite_foods.values())
ax[1, 0].set_title("Favorite Foods")

# Histogram of 1000 random normally distributed numbers on the bottom right subplot
X_norm = np.random.randn(1000)
ax[1, 1].hist(X_norm, bins=30, edgecolor='black')
ax[1, 1].set_title("Normal Distribution Histogram")

plt.tight_layout()
plt.show()
```
